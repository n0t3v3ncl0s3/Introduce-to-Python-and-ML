# Singular Value Decomposition

```python
import numpy as np
import random
from  numpy import linalg as la
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
random.seed(42)
img = mpimg.imread('image2.jpg')
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
img.shape
```

```python
img = img.mean(axis=2)
U, S, V = la.svd(img)
```

```python
def truncated(U,S,V,k):
    return U[:,:k] @ np.diag(S[:k]) @ V[:k,:]
```

```python
for k in [2,5,10,100,200]:
    plt.imshow(truncated(U,S,V,k),cmap='gray')
    plt.title(f'k = {k}')
    plt.axis('off')
    plt.show()
```
